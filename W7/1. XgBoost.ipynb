```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Tugas Matkul Machine Learning/W7/Dataset di W7/train.csv')
data.head()
```

```python
data.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = data.SalePrice
X = data.drop(['SalePrice'], axis=1).select_dtypes(exclude=['object'])
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.25)

my_imputer = SimpleImputer()
train_X = my_imputer.fit_transform(train_X)
test_X = my_imputer.transform(test_X)
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor(n_estimators=15000, learning_rate=0.015, max_depth=5,num_parallel_tree=5 )
my_model.fit(train_X,train_y,early_stopping_rounds=10,eval_set=[(test_X,test_y)],verbose=False)
```

```python
preds = my_model.predict(test_X)
from sklearn.metrics import mean_absolute_error
print('Mean Absolute Error: ' + str(mean_absolute_error(preds, test_y)))
```
